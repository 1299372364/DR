```python
from sklearn.multiclass import OneVsRestClassifier
import pickle
import numpy as np
from sklearn.svm import SVC
from sklearn import decomposition 
import os
import pandas as pd
from PIL import Image
from scipy import interp
from sklearn import model_selection
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import label_binarize,scale
from sklearn import manifold
from sklearn.decomposition import PCA
import seaborn as sns
import time
import quadratic_weighted_kappa as qwk
import Chinese
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#生成文件名列表
```

```python
start=time.clock()
FindPath="ImageOperate/images/"
filenames=os.listdir(FindPath)
imlist=[]
for name in filenames:
    filePath=os.path.join(FindPath,name)
    imlist.append(filePath)
```

```python
data = np.array([np.array(Image.open(im)).flatten() for im in imlist],'f')#载入图像数据并将其化为shape(1,262144)
```

```python
data.shape
```

```python
x=data
```

```python
y=pd.read_csv("trainLabels4.csv",index_col='image')#载入标记
```

```python
n_features=262144
```

```python
h=w = 512
```

```python
n_samples=2250
```

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(x, y, test_size=0.3,
                                                                    stratify=y, random_state=42)
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

#接下来做PCA降维

```python
n_components = 150#目标是下降到150维
print("Extracting the top %d eigenDR from %d DR"
      % (n_components, X_train.shape[0]))
pca = PCA(n_components=n_components, svd_solver='randomized',whiten=True).fit(X_train)
eigenDR =pca.components_.reshape((n_components,h, w))
print("Projecting the input data on the eigenDR orthonormal basis")
X_train_pca = pca.transform(X_train)
X_test_pca = pca.transform(X_test)
X_train_pca.shape
X_test_pca.shape
```

```python
X_train_pca.shape
```

```python
print("Fitting the classifier to the training set")
classifier = OneVsRestClassifier(SVC(kernel='rbf',class_weight='balanced',C=1))
classifier.fit(X_train_pca, y_train)#模型拟合
#with open("save_model/CNN_SVM_classifier.pickle",'wb') as f:
    #pickle.dump(classifier,f)#储存分类器
#with open("save_model/CNN_SVM_classifier.pickle",'rb') as f:
    #classifier=pickle.load(f)#恢复分类器
y_pred = classifier.predict(X_test_pca)#预测
print(classification_report(y_test, y_pred,target_names=('No DR', 'Mild', 'Moderate', 'Severe', 'Proliferative')))
print(confusion_matrix(y_test,y_pred,labels=range(5)))
```

```python
labels = [0,1,2,3,4]
cm = confusion_matrix(y_test, y_pred, labels=labels)
print(cm)
```
